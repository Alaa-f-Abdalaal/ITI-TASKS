```python
import pandas as pd
avg_price=df["price"].mean
print("Average Price:", avg_price)
max_price = df["price"].max()

print("Max Price:", max_price)
```

```python
avg_price, avg_num_items = df[["price", "num_item"]].mean()
print("Avg Price:", avg_price)
print("Avg Num Items:", avg_num_items)
```

```python
df["total"] = df["price"] * df["num_item"]
print(df.head())
```

```python
avg_top3 = df.nlargest(3, "price")["price"].mean()
print("Avg price of top 3 expensive purchases:", avg_top3)
```

```python
num_products = df["product_id"].nunique()
print("Number of different products:", num_products)
```

```python
stats_total = df["total"].describe()
diff_mean_median = stats_total["mean"] - stats_total["50%"]
print("Difference:", diff_mean_median)
```

```python
df.drop("total", axis=1, inplace=True)
print(df.head())
```

```python
df["invoice_time"] = pd.to_datetime(df["invoice_time"])
print(df.dtypes)
```

```python
df["DOW"] = df["invoice_time"].dt.day_name()
df["Month"] = df["invoice_time"].dt.month_name()
print(df.head())
```

```python
num_purchases_wed = df["DOW"].value_counts()["Wednesday"]
print("Purchases on Wednesday:", num_purchases_wed)
```

```python
missing_age = df["Age"].isnull().sum()
print("Missing Age entries:", missing_age)
```

```python
avg_age = df["Age"].mean()
print("Average Age:", avg_age)
```

```python
df["Age"].fillna(avg_age, inplace=True)
```

```python
avg_5_oldest = df.sort_values("Age", ascending=False).head(5)["Age"].mean()
print("Average age of 5 oldest passengers:", avg_5_oldest)
```
