```python
# --------------------------
# Introduction & Setup
# --------------------------
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

# --------------------------
# Basic Matplotlib
# --------------------------


plt.plot([1, 2, 3, 4, 5])
plt.title("Simple Line Plot")
plt.show()

# X & y plot
X = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]
plt.plot(X, y)
plt.title("X vs y Line Plot")
plt.xlabel("X")
plt.ylabel("y")
plt.show()

# OO method plot
fig, ax = plt.subplots()
ax.plot(X, y)
ax.set_title("OO Method Plot")
ax.set_xlabel("X")
ax.set_ylabel("y")
plt.show()

# Matplotlib workflow
fig, ax = plt.subplots()
ax.plot(X, y)
ax.set_title("Workflow Example")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
fig.savefig("workflow_plot.png")
plt.show()


# --------------------------
# Using NumPy with Matplotlib
# --------------------------

# X vs X^2
X = np.linspace(0, 100, 100)
fig, ax = plt.subplots()
ax.plot(X, X**2)
ax.set_title("X vs X^2")
plt.show()

# Scatter plots (exp)
plt.scatter(X, np.exp(X/20))  # scale to avoid huge numbers
plt.title("Scatter: X vs exp(X)")
plt.show()

# Scatter plots (sin)
plt.scatter(X, np.sin(X))
plt.title("Scatter: X vs sin(X)")
plt.show()


# --------------------------
# Bar Plots
# --------------------------

# Vertical bar
foods = {"Pizza": 50, "Burger": 30, "Sushi": 70}
plt.bar(foods.keys(), foods.values())
plt.title("Favourite Foods (Vertical)")
plt.xlabel("Food")
plt.ylabel("Price")
plt.show()

# Horizontal bar
plt.barh(list(foods.keys()), list(foods.values()))
plt.title("Favourite Foods (Horizontal)")
plt.xlabel("Price")
plt.ylabel("Food")
plt.show()


# --------------------------
# Histograms
# --------------------------

# Normal distribution
data_normal = np.random.randn(1000)
plt.hist(data_normal, bins=30)
plt.title("Normal Distribution Histogram")
plt.show()

# Uniform distribution
data_uniform = np.random.rand(1000)
plt.hist(data_uniform, bins=30)
plt.title("Uniform Distribution Histogram")
plt.show()


# --------------------------
# Subplots
# --------------------------

# 2x2 subplot
fig, axes = plt.subplots(2, 2, figsize=(10, 5))

# Top left: line plot
axes[0,0].plot(X, X/2)
axes[0,0].set_title("X vs X/2")

# Top right: scatter
axes[0,1].scatter(np.random.rand(10), np.random.rand(10))
axes[0,1].set_title("Random Scatter")

# Bottom left: bar
axes[1,0].bar(foods.keys(), foods.values())
axes[1,0].set_title("Favourite Foods")

# Bottom right: histogram
axes[1,1].hist(np.random.randn(1000), bins=30)
axes[1,1].set_title("Normal Dist Histogram")

plt.tight_layout()
plt.show()


# --------------------------
# Pandas + Matplotlib
# --------------------------

# Load dataset
car_sales = pd.read_csv("/content/car-sales.csv")

# Clean Price (remove $ and convert to float)
car_sales["Price"] = car_sales["Price"].str.replace("[\$,]", "", regex=True).astype(float)

# Add Total Sales & Sale Date
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales["Sale Date"] = pd.date_range(start=pd.Timestamp.today().normalize(), periods=len(car_sales))

# Line plot (Total Sales vs Date)
car_sales.plot(x="Sale Date", y="Total Sales", kind="line", title="Total Sales Over Time")
plt.show()

# Scatter (Odometer vs Price)
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter", title="Odometer vs Price")
plt.show()

# Random DataFrame bar plot
X = np.random.rand(10, 4)
df = pd.DataFrame(X, columns=["a", "b", "c", "d"])
df.plot(kind="bar", title="Random Data Bar Plot")
plt.show()

# Bar plot (Make vs Odometer)
car_sales.plot(x="Make", y="Odometer (KM)", kind="bar", title="Make vs Odometer")
plt.show()

# Histogram (Odometer)
car_sales["Odometer (KM)"].plot(kind="hist", title="Histogram of Odometer (KM)")
plt.show()

# Histogram (Price with bins)
car_sales["Price"].plot(kind="hist", bins=20, title="Histogram of Price")
plt.show()
```
